```python
from PIL import Image
import os, sys
import cv2
import numpy as np
import pandas as pd
import random
import tensorflow as tf
from tensorflow import keras 
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential,Model, load_model
from keras.preprocessing.image import ImageDataGenerator, load_img
import math
from retinaface import RetinaFace
```

```python
def pred_result(path):
  plt.figure(figsize=(5,5))
  image_index = os.path.basename(path)
  img = load_img(path)
  img = img.resize((224,224),Image.ANTIALIAS)
  arr = np.array([np.array(img)])
  pre1 = np.argmax(pre.predict(arr))
  plt.imshow(img)
  plt.title(f"predict {label[pre1]} - {image_index.split('_')[0]}")
```

```python
label = ["Đức","Hùng","Quân","Trường","Việt Đức","HĐức","Kiên","Tân","Tuấn","Xuân Anh","Hiếu","Linh","Thắng","Vân"]
```

```python
pre = tf.keras.models.load_model('/content/drive/MyDrive/Colab Notebooks/model.h5')
```

```python
def pred_result_detect(path):
  plt.figure(figsize=(5,5))
  image_index = os.path.basename(path)
  img = load_img(path)
  resp = RetinaFace.detect_faces(path)
  faces = RetinaFace.extract_faces(img_path = path, align = True)
  for face in faces:
    face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
    cv2.imwrite("filename1.jpg", face)
    img1 = load_img("filename1.jpg")
    img1 = img1.resize((224,224),Image.ANTIALIAS)
    arr = np.array([np.array(img1)])
    pre1 = np.argmax(pre.predict(arr))
  plt.imshow(img)
  plt.title(f"predict {label[pre1]} - {image_index.split('_')[0]}")
```
